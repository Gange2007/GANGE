```python
import plotly.express as px
import numpy as np
import pandas as pd
time=np.linspace(0,10,100)
voltage=np.sin(time)
df=pd.DataFrame({'time':time,'voltage':voltage})
fig=px.line(df,x='time',y='voltage',title='AC Voltage Signal')
fig.show()
```
